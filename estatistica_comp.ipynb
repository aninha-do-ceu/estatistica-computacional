```python
import random

def simular_markov(P, estado_inicial=1, n_iter=10000, burn_in=1000):
    cadeia_i = []
    estado = estado_inicial
    visitas_estado1 = 0
    total_util = 0

    for t in range(n_iter):
        # Registrar visitas após burn-in
        if t >= burn_in:
            if estado == 1:
                visitas_estado1 += 1
            total_util += 1
        
        # Gerar próximo estado
        u = random.random()
        if estado == 1:
            estado = 1 if u < P[0][0] else 2
        else:  # estado = 2
            estado = 1 if u < P[1][0] else 2
        cadeia_i.append(estado)

    # Estimativa da probabilidade invariante de estar no estado 1
    pi1_est = visitas_estado1 / total_util
    return pi1_est,cadeia_i

# matriz de transição
P = [[0.75, 0.25], 
     [0.125, 0.875]]

pi1,traco_cadeia = simular_markov(P, estado_inicial=1, n_iter=50000, burn_in=5000)
print("Estimativa da probabilidade invariante de estado 1:", pi1)
print("Estimativa da probabilidade invariante de estado 2:", 1 - pi1)
```

```python
# alvo: N(0,1). Proposta 1: N(x,0,5)

n_iter=10000
sigmas = [0.5,0.1,10]

dict_cadeias = {}

for sigma in sigmas:
    x_inicial = 2
    x_atual = x_inicial
    cadeia = []
    aceitos = 0
    for i in range(n_iter):
        # gere y de N(x,0,5) com x atual
        y = st.norm.rvs(loc=x_atual, scale= sigma)

        # calcule prob de aceitacao
        prob = min(1, np.exp(-(y**2 - x_atual**2)/2))

        # gere u de uniforme(0,1) e veja se a prob de aceitacao é maior
        u = st.uniform.rvs()

        if u <= prob:
            cadeia.append(y)
            x_atual = y
            aceitos += 1
        else:
            x_atual = x_atual
            
        dict_cadeias[sigma] = [aceitos, cadeia] 
```

```python
for sigma in sigmas:
    print(f'Para sigma = {sigma} temos taxa de aceitação: {dict_cadeias[sigma][0]/n_iter}\n')
```

```python
fig, axs = plt.subplots(ncols = 3, nrows = 3, figsize = (12,8))

cont = 0
for sigma in sigmas:
    
    axs[cont,0].plot(dict_cadeias[sigma][1])
    axs[cont,0].set_title(f"Traço das cadeias - Sigma = {sigma}")
    
    plot_acf(dict_cadeias[sigma][1], lags=50, ax=axs[cont,1])
    
    sns.kdeplot(dict_cadeias[sigma][1], shade=True,ax=axs[cont,2])
    axs[cont,2].set_title(f"Densidade da Normal Alvo")
    
    cont += 1
    
plt.tight_layout()
```

## Cadeias de Markov

A distribuição alvo é $f(x)$ e a proposta é $g(x)$. 
Algoritmo:

1. Gere $y \sim g(x)$.
2. Calcule a probabilidade de aceitação: $p(x,y)$

$$
p(x,y) = min(1,\frac{f(y)g(x|y)}{f(x)g(y|x)})
$$

3. Gere $u \sim U(0,1)$. Se $u \leq p(x,y)$, aceite o valor e faça $x^{(t+1)} = y$, se não, $x^{(t+1)} = x^{(t)}$

4. Incremente o contador de $t$ para $t+1$ e volte ao passo 1.

## Amostrador Independente

É quando a distribuição proposta não depende do estado atual da cadeia, ou seja, $g(y|x) = g(y)$. Logo, a probabilidade de aceitação fica $p(x,y) = min(1, \frac{f(y)g(x)}{f(x)g(y)})$.

### Teorema

O amostrador independente produz cadeia ergódica se existe $M$ tal que 

$$
\frac{f(x)}{q(x)} \leq M, \any x
$$

Assim, a probabilidade esperada de aceitação é no mínimo $1/M$ se a cadeia for estacionária. 


## Algoritmo de Metropolis

O algoritmo considera apenas propostas simétricas, ou seja, $g(x|y) = g(y|x)$ e, assim, a probabilidade de aceitação fica $p(x,y) = min(1,\frac{f(y)}{f(x)})$.
Uma proposta comum é $g(y|x) = g(|x-y|)$, porém deve-se analisar diversos valores de $\sigma^2$, pois, se for muito pequeno, muitos valores serão aceitos e vai precisar de muitas iterações para cobrir todo o espaço de $X$, do contrário, muitos valores serão rejeitados e a cadeia se movimenta muito pouco.



## Amostragem em Blocos

Seja $\mathbf{x} = (x_1,...,x_d)'$ vetor de parâmetros de dimensão $d$.
Nesse caso, pode ser melhor dividir $\mathbf{x}$ em $k$ blocos ${\mathbf{x}_1,...,\mathbf{x}_k}$ e dentro de cada iteração teremos o algoritmo aplicado $k$ vezes.

O vetor $\mathbf{x}_{-i} = (\mathbf{x}_1,...,\mathbf{x}_{i-1},\mathbf{x}_{i+1},...,\mathbf{x}_k)$ é o vetor que contém todos os elementos de $\mathbf{x}$ exceto $\mathbf{x}_i$.

Suponha que na iteração $t+1$ os blocos $1,2...,i-1$ foram atualizados. Para atualizar a i-ésima componente, um valor de $\mathbf{x}_i$ é gerado a partir da proposta $g(|\mathbf{x}_i,\mathbf{x}_{-i})$ e o valor é aceito com probabilidade

$$
p(\mathbf{x}_i,\mathbf{x}'_i) = min(1, \frac{f(\mathbf{x}'_i|\mathbf{x}_{-i}) g(\mathbf{x}_i|\mathbf{x}'_i,\mathbf{x}_{-i})}{f(\mathbf{x}_i|\mathbf{x}_{-i}) g(\mathbf{x}'_i|\mathbf{x}_i,\mathbf{x}_{-i})})
$$

Onde $f(\mathbf{x}_i|\mathbf{x}_{-i})$ é a distribuição condicional completa de $\mathbf{x}_i$.

### Algoritmo MALA 

Para um vetor aleatório $\mathbf{x} \in R^d$ com densidade $f(\mathbf{x})$ a versão em tempo discreto é

$$
x^{t+1} = x^t + \frac{\sigma^2}{2} \nabla_{\mathbf{x}} log f(\mathbf{x}^{(t)}) + \sigma \epsilon_t 
$$

Com $\epsilon_t \sim N(0,I_d)$.

Assim,

$$
\mathbf{X}^{(t+1)}|\mathbf{X}^{(t)} \sim N(\mathbf{x}^{(t)} + \frac{\sigma^2}{2} \nabla_{\mathbf{x}} log f(\mathbf{x}^{(t)}), \sigma^2 I_d)
$$
