```python
# taxa SELIC e índice IBOV

selic = pd.read_csv('taxa_juros_selic.csv', sep=";")
ibov = pd.read_csv('ibov.csv',sep=",")
```

```python
selic.head()
```

```python
ibov.head()
```

```python
selic.columns = ['data','selic']
ibov.columns = ['indice','data','ibov']
```

```python
selic = selic[selic['selic'] != 'BCB-Demab']
selic['selic'] = selic['selic'].str.replace(',', '.').astype(float)
selic['data'] = pd.to_datetime(selic['data'], format='%d/%m/%Y')
selic.head()
```

```python
ibov = ibov[['data','ibov']]
ibov['data'] = pd.to_datetime(ibov['data'], format='%d.%m.%Y')
ibov['ibov'] = ibov['ibov'].astype(float)
ibov.head()
```

```python
print(min(ibov['data']))
print(min(selic['data']))
```

```python
df_series = ibov.merge(selic, on = 'data', how = 'inner')
df_series.head()
```
